```python
#library
import tensorflow as tf
import keras
from keras import layers
```

```python
#data
x=tf.constant([1,-1,0,2,-2,3,-3,4,-4,5,6],dtype=float)
y=x*3-5
print("x=",x.numpy(),"\ny=",y.numpy())
```

```python
model=keras.Sequential([
    layers.Input(shape=(1,)),
    layers.Dense(units=1,activation=None)
])
```

```python
model.compile(optimizer='sgd',loss='mean_squared_error')
model.summary()
```

```python
model.fit(x,y,batch_size=1,epochs=400)
```

```python
x_test=tf.constant([-6,-5,7,8,9,10,-10],dtype=float)
y_test=x_test*3-5
print("x_test=",x_test.numpy(),"\ny_test=",y_test.numpy())
```

```python
y_pred=model.predict(x_test)
print("y_pred=",y_pred)
```
